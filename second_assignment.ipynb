(1)-Discuss string slicing and provide examples??
---------------------------------------------------------------------

Ans-
String slicing in Python allows you to extract a portion of a string by specifying a start and end index. It follows the format string[start:end:step], where start is the index where slicing begins (inclusive) and end is the index where it ends (exclusive).

Syntax:

string[start:stop:step]

Examples:

(1) s = "Hello, World!"

print(s[0:5])  # Output: Hello

* This extracts the substring from index 0 to 4 (the 5 is exclusive).

(2) s = "Hello, World!"

print(s[:5])   # Output: Hello (same as s[0:5])

print(s[7:])   # Output: World! (from index 7 to the end)

* When start is omitted, it starts from the beginning. When stop is omitted, it goes to the end.

(3) s = "Hello, World!"

print(s[-6:-1])  # Output: World

* Negative indices count from the end of the string. Here, -6 refers to the character W and -1 refers to ! (not included).

(4) s = "Hello, World!"

print(s[::2])  # Output: Hlo ol!

* This extracts every second character from the string.

(2)-  Explain the key  features of  lists in Python 
-------------------------------------------------------------------

Ans-Lists in Python are mutable sequences used to store collections of items. Here are the key features of lists in Python:

1. Ordered Collection:

* Lists maintain the order of elements. The order in which elements are added is preserved.

 Examples: 
 
 fruits = ['apple', 'banana', 'cherry']
 
print(fruits)  # Output: ['apple', 'banana', 'cherry']

2. Mutability:

* Lists are mutable, meaning their elements can be changed after the list has been created.

 Examples: 
 
 fruits = ['apple', 'banana', 'cherry']
 
fruits[1] = 'blueberry'

print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

3. Heterogeneous Elements:

* Lists can contain elements of different data types, including integers, strings, floats, and even other lists.

Examples: 

mixed_list = [1, 'hello', 3.14, [5, 6, 7]]

print(mixed_list)  # Output: [1, 'hello', 3.14, [5, 6, 7]]


4. Indexing and Slicing:

* Elements in a list can be accessed using indexing and slicing, similar to strings.

Example:

letters = ['a', 'b', 'c', 'd']

print(letters[1])   # Output: b

print(letters[1:3]) # Output: ['b', 'c']

5. Iteration:

* Lists can be easily iterated over using loops.

Example:

for fruit in ['apple', 'banana', 'cherry']:
    print(fruit)
    
Output:

* apple
* banana
* cherry

(3)-Describe how to access, modify and delete elements in a list with examples?? 
---------------------------------------------------------------------------------------------------------------------

Ans 
* Accessing Elements in a List:

(1) By Index:

Elements in a list can be accessed using their index. The index is zero-based, meaning the first element is at index 0.

Examples:

fruits = ['apple', 'banana', 'cherry']

print(fruits[0])  # Output: apple

print(fruits[2])  # Output: cherry

(2) Using Slicing:

Slicing allows you to access a range of elements.

Examples:

fruits = ['apple', 'banana', 'cherry', 'date']

print(fruits[1:3])  # Output: ['banana', 'cherry']

print(fruits[:2])   # Output: ['apple', 'banana']

print(fruits[2:])   # Output: ['cherry', 'date']

print(fruits[::2])  # Output: ['apple', 'cherry']


* Modifying Elements in a List:

Changing an Element by Index You can change the value of an element by specifying its index.

Examples:

fruits = ['apple', 'banana', 'cherry']

fruits[1] = 'blueberry'

print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

* Modifying by using Slice:

You can replace multiple elements at once by assigning to a slice.

Examples:

fruits = ['apple', 'banana', 'cherry', 'date']

fruits[1:3] = ['blueberry', 'blackberry']

print(fruits)  # Output: ['apple', 'blueberry', 'blackberry', 'date']


* Deleting Elements in a List:

Using del Statement You can delete an element or a slice using the del statement.

Examples:

fruits = ['apple', 'banana', 'cherry']

del fruits[1]

print(fruits)  # Output: ['apple', 'cherry']

* Deleting a slice

fruits = ['apple', 'banana', 'cherry', 'date']

del fruits[1:3]

print(fruits)  # Output: ['apple', 'date']


* Using remove() Method:

The remove() method deletes the first occurrence of a value.

Examples:

fruits = ['apple', 'banana', 'cherry', 'banana']

fruits.remove('banana')

print(fruits)  # Output: ['apple', 'cherry', 'banana']

* Using pop() Method
The pop() method removes and returns the element at a specified index. If no index is specified, it removes and returns the last element.

Examples:

fruits = ['apple', 'banana', 'cherry']

popped_fruit = fruits.pop(1)

print(popped_fruit)  # Output: banana

print(fruits)        # Output: ['apple', 'cherry']

* Popping the last element

popped_fruit = fruits.pop()

print(popped_fruit)  # Output: cherry

print(fruits)        # Output: ['apple']

Lists: Defined using square brackets [].

* Tuples are ideal for fixed collections of items, structured data records, function arguments and return values, and situations where immutability is important.

* Sets are best for collections of unique items, efficient membership testing, performing set operations, and situations where you need to ensure data uniqueness or handle graph-related problems.

location_data = {

    ("New York", "USA"): 8175133,
    
    ("Los Angeles", "USA"): 3792621
}

* Structured Data Records

Simple Data Records: Storing related pieces of data together.

person = ("John", "Doe", 30)


* Use Cases of Sets:


* Unique Collections of Items

Removing Duplicates: Quickly removing duplicates from a list.

nums_list = [1, 2, 2, 3, 4, 4, 5]

unique_nums = list(set(nums_list))

print(unique_nums)  # Output: [1, 2, 3, 4, 5]

Fast Membership Testing: Sets provide an efficient way to check if an item is in a collection.

allowed_users = {"alice", "bob", "charlie"}

if "dave" in allowed_users:

    print("Access granted")
    
else:

    print("Access denied")


* Set Operations

Union, Intersection, Difference: Performing standard set operations.

set1 = {1, 2, 3}

set2 = {3, 4, 5}

union = set1 | set2

print(union)  # Output: {1, 2, 3, 4, 5}

intersection = set1 & set2

print(intersection)  # Output: {3}

difference = set1 - set2

print(difference)  # Output: {1, 2}

(8) Discuss the importance of  dictionary keys being immutable and provide examples??
---------------------------------------------------------------------------------------------------------------------------

Ans-  Python, dictionary keys must be immutable. This immutability is essential for maintaining the integrity and performance of dictionaries. 

* Importance of Immutable Dictionary Keys:

Once a key is created, it cannot be changed. Immutability is essential for keys because it ensures that the dictionary can efficiently look up values based on their keys. If a key was mutable, its hash value could change, making it impossible to find the associated value in the dictionary.

* Examples of Immutable Keys:

* Using Strings as Keys
Strings are immutable and commonly used as dictionary keys.

student_grades = {'Alice': 'A', 'Bob': 'B', 'Charlie': 'C'}

print(student_grades['Alice'])  # Output: A

* Using Numbers as Keys

Numbers (integers and floats) are immutable and can be used as dictionary keys.

squares = {1: 1, 2: 4, 3: 9, 4: 16}

print(squares[3])  # Output: 9

* Using Tuples as Keys

Tuples are immutable and can be used as dictionary keys, especially useful for composite keys.

coordinates = {(0, 0): 'origin', (1, 2): 'point A', (3, 4): 'point B'}

print(coordinates[(1, 2)])  # Output: point A

* Examples of Mutable Objects as Keys (Not Allowed)

* Using dictionary as Keys:


Dictionaries themselves are mutable and cannot be used as keys.
